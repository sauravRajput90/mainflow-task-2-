# Data Manipulation with pandas

```python
# Import The Library
import pandas as pd 
```

```python
# Reading The Dataset
df = pd.read_csv('E:\postings.csv')
```

```python
type(df)
```

```python
# concise summary 
df.info
```

```python
# Generate Descriptive Statistics
df.describe()
```

```python
# remove duplicate rows
df=df.drop_duplicates()
df
```

```python
# detect missing values
df.isnull()
```

```python
# total number of missing values 
df.isnull().sum()
```

```python
# detects existing (non-missing) values
df.notnull()
```

```python
#  total number of missing values in the entire DataFrame
df.isnull().sum().sum()
```

```python
# replace missing values (NaN)
df2=df.fillna(0)
df2
```

```python
# fill missing values (NaN)
# it replaces it with the value from the previous row that is not null
df3=df.fillna(method='pad')
df3
```

```python
# fill missing values (NaN)
# it replaces it with the value from the next row that is not null
df4=df.fillna(method='bfill')
df4
```
